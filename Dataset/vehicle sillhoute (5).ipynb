<h3>1.1:Information Gain</h3>

Information gain is a key concept in the field of machine learning and
decision trees. It is a metric used to measure the effectiveness of an
attribute in classifying data. In the context of decision trees,
information gain helps determine the most valuable attribute for
splitting data by quantifying the amount of uncertainty reduced after
splitting the data based on a particular attribute. In essence, it aids
in identifying the attributes that contribute the most to the
classification process. In this R code, we'll explore the calculation
and application of information gain.

<h5>Some information to Know</h5>

We aim to conduct three different data splits using the same formula to
observe the variations in outcomes. After exploring numerous attribute
combinations to construct the formula based on information gain, we
settled on utilizing (scaled_variance.1, scatter_ratio, elongatedness,
scaled_variance, radius_ratio) as our selected attributes. This decision
was made after testing multiple attribute combinations and finding that
these particular attributes consistently led to clear and non-confusing
tree representations. Employing the same formula across the three splits
will enable us to explicitly observe the distinctions arising from the
different datasets and how these variations affect the decision tree
structures.

<h5>Steps we will apply here:</h5>

In this selection process, we initiated by dividing the dataset into
training and testing subsets across three varied partitions. A uniform
formula was then employed consistently across all three partitions.
Subsequently, decision trees were constructed using the training set,
and predictions were evaluated, followed by the visualization of the
tree structure. The model was utilized to predict class labels for the
test data, and a thorough evaluation, including the extraction of
measures such as accuracy, was conducted to assess the effectiveness of
the partitioning strategy.

<h4>1.1.A:split on 70% training and 30%testing</h4>

<h6>First we split dataset into training and testing subnets</h6>


```r
##Needed library
library(party)
library(e1071)
library(caret)
```

```r
# 1. Split the dataset into training and testing subsets
set.seed(123)  # Set a seed for reproducibility
ind <- sample(2, nrow(dataset), replace = TRUE, prob = c(0.7, 0.3))
train.data <- dataset[ind == 1, ]
test.data <- dataset[ind == 2, ]
```

```r
# 2. Define the formula for the model
myFormula <- Class ~ scaled_variance.1  + scatter_ratio + elongatedness + scaled_variance + radius_ratio
```

```r
# 3. Build the decision tree using the training set
dataset.ctree <- ctree(myFormula, data = train.data)
```

```r
# 4. Check the prediction on the training set
trainPred <- predict(dataset.ctree)
table(trainPred, train.data$Class)
```

```r
# 7. Evaluate the model
library(e1071)
library(caret)
co_result <- confusionMatrix(result)
print(co_result)


# Extracting sensitivity and specificity from the table in co_result
table_matrix <- co_result$table

# Extracting true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
TP <- table_matrix[2, 2]  # True Positive
TN <- sum(diag(table_matrix)) - sum(table_matrix[2, ]) - sum(table_matrix[, 2]) + TP  # True Negative
FP <- sum(table_matrix[, 2]) - TP  # False Positive
FN <- sum(table_matrix[2, ]) - TP  # False Negative

# Calculate sensitivity and specificity
sensitivity <- TP / (TP + FN)
specificity <- TN / (TN + FP)

# Print the sensitivity and specificity
print(sensitivity)
print(specificity)




accuracy <- co_result$overall["Accuracy"]
accuracy
```
