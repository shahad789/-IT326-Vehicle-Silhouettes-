<h3>1.1:Information Gain</h3>

Information gain is a key concept in the field of machine learning and
decision trees. It is a metric used to measure the effectiveness of an
attribute in classifying data. In the context of decision trees,
information gain helps determine the most valuable attribute for
splitting data by quantifying the amount of uncertainty reduced after
splitting the data based on a particular attribute. In essence, it aids
in identifying the attributes that contribute the most to the
classification process. In this R code, we'll explore the calculation
and application of information gain.

<h5>Some information to Know</h5>

We aim to conduct three different data splits using the same formula to
observe the variations in outcomes. After exploring numerous attribute
combinations to construct the formula based on information gain, we
settled on utilizing (scaled_variance.1, scatter_ratio, elongatedness,
scaled_variance, radius_ratio) as our selected attributes. This decision
was made after testing multiple attribute combinations and finding that
these particular attributes consistently led to clear and non-confusing
tree representations. Employing the same formula across the three splits
will enable us to explicitly observe the distinctions arising from the
different datasets and how these variations affect the decision tree
structures.

```python
##Needed library
library(party)
library(e1071)
library(caret)
```

```python
# 1. Split the dataset into training and testing subsets
set.seed(123)  # Set a seed for reproducibility
ind <- sample(2, nrow(dataset), replace = TRUE, prob = c(0.7, 0.3))
train.data <- dataset[ind == 1, ]
test.data <- dataset[ind == 2, ]
```

```python
# 2. Define the formula for the model
myFormula <- Class ~ scaled_variance.1  + scatter_ratio + elongatedness + scaled_variance + radius_ratio
```

```python
# 3. Build the decision tree using the training set
dataset.ctree <- ctree(myFormula, data = train.data)
```

```python
# 4. Check the prediction on the training set
trainPred <- predict(dataset.ctree)
table(trainPred, train.data$Class)
```

```python
# 7. Evaluate the model
library(e1071)
library(caret)
co_result <- confusionMatrix(result)
print(co_result)


# Extracting sensitivity and specificity from the table in co_result
table_matrix <- co_result$table

# Extracting true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN)
TP <- table_matrix[2, 2]  # True Positive
TN <- sum(diag(table_matrix)) - sum(table_matrix[2, ]) - sum(table_matrix[, 2]) + TP  # True Negative
FP <- sum(table_matrix[, 2]) - TP  # False Positive
FN <- sum(table_matrix[2, ]) - TP  # False Negative

# Calculate sensitivity and specificity
sensitivity <- TP / (TP + FN)
specificity <- TN / (TN + FP)

# Print the sensitivity and specificity
print(sensitivity)
print(specificity)




accuracy <- co_result$overall["Accuracy"]
accuracy
```

<p>Clustering, an unsupervised machine learning technique, involves organizing a collection of data points into distinct subsets or "clusters." The aim is to unite similar data points within each cluster while keeping them distinct from those in other clusters. In this method, we'll experiment with three different 'k' sizes (number of clusters). Subsequently, we'll employ evaluation techniques to compare these various 'k' sizes and determine which one best suits our dataset.For every different size of k we will try a different method of clustering.</p>

We'll initiate the clustering process by selecting an initial k size to apply in k-means three times. We begin with k=3, representing our class labels. Next, we increment k by 1 (k=4) and then decrement it by 1 to observe how it influences the clusters, building on our initial choice of k=3 based on class labels. Subsequently, we calculate precision and recall for all scenarios and visualize the clusters to assess the correctness of our choices. Finally, we apply various evaluation techniques to analyze and determine the optimal number of clusters.

<h4>Calculation of Precision and recall</h4>
In the forthcoming code sections, we will compute precision and recall for each class within every cluster. Initially, we determine the size of each cluster. Following that, we utilize the code result = -print(sum(datacluster$Class[kmeanResult$cluster == "1"] == "1", na.rm = TRUE)) to find the count of instances for each class (whether it's 1, 2, or 3) in a specific cluster. This process is applied to all classes based on the number of k-clusters.

The outcome obtained from this code is then incorporated into the precision and recall formulas. Precision is determined by taking the count from the code and dividing it by the total size of the individual cluster being calculated. For example, when calculating for k=3 and class 1 in cluster one, we use the result from the code to find the count of class 1 in cluster one. This count is divided by the first size obtained from kmeanResults$size representing cluster 1, and the result is multiplied by 100 to express it as a percentage.

For calculating recall, the same result from the code for an individual class in a specific cluster is taken, but this time it is divided by the total number of instances of that class in our dataset. This calculation approach is applied across different sizes of k and for all classes within various clusters.
